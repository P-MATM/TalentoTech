# Agrupar (filtrar) por una categoría (columna): Agrupo, para ver el resumen (valor agrupado, total)  de ventas por región.

*   df.groupby()[].sum()

 (categoría: columna)[varible numérica].sum()

____________________________________________


```python
# Ejercicio 2: Calcular el total de ventas por región
print("Ejercicio 2: Calcular el total de ventas por región")
ventas_por_region = df.groupby('Region')['Ventas'].sum()
print(ventas_por_region)
print("\n")
print(df.head(18))
```

```python
# Ejercicio 3: Filtrar el Dataframe para mostrar sólo las ventas realizadas por David
print("Ejercicio 3: Filtrar el Dataframe para mostrar sólo las ventas realizadas por David")
ventas_david = df[df['Vendedor'] == 'David']
print(ventas_david)
print("\n")
```

```python
# Ejercicio 4: Calcular el promedio de ventas por producto
print("Ejercicio 4: Calcular el promedio de ventas por producto")
ventas_por_producto = df.groupby('Producto')['Ventas'].mean()
print(ventas_por_producto)
print("\n")
```

```python
# Ejercicio 5: Ordenar el Dataframe por ventas, en orden descendente
print("Ejercicio 5: Ordenar el Dataframe por ventas en orden descendente")
df_ordenado = df.sort_values(by='Ventas',ascending=False)
print(df_ordenado)
```

# Manipulación de datos:


*   Filtrar las filas donde las ventas sean mayores a 10.000.
*   Ordena el DataFrame por la columna "Ventas" en orden descendente.
*   Elimina la columna "Vendedor".

___________________________________________





```python
# Ejercicio 6: Filtrar las filas con ventas mayores a 10.000
print('Ejercicio 4: Filtrar las filas con ventas mayores a 10.000')
df_filtrado= df[df['Ventas']>10000]
print(df_filtrado)
print("\n")
```

```python
# Ejercicio 7: Ordenar el DataFrame por ventas en orden descendente
print('Ordenar el DataFrame por ventas en orden descendente')
df_ordenado = df.sort_values(by='Ventas',ascending=False)
print(df_ordenado)
print("\n")

## Ejercicio 7.1: Ordenar el DataFrame por ventas en orden ascendente
print("Ordenar el DataFrame por ventas en orden ascendente")
df_ordenado = df.sort_values(by='Ventas',ascending=True)
print(df_ordenado)
print("\n")

## Ejercicio 7.2: Ordenar el DataFrame por ventas en orden ascendente, sin usar ascending
print("Ordenar el DataFrame por ventas en orden ascendente, sin usar ascending")
df_ordenado = df.sort_values(by='Ventas',ascending=True)
print(df_ordenado)
print("\n")
```

```python
# Ejercicio 8: Eliminar la columna "Vendedor"
print('Eliminar la columna "Vendedor"')
df_sin_vendedor = df.drop(columns=['Vendedor'])
print(df_sin_vendedor)
print("\n")
```
